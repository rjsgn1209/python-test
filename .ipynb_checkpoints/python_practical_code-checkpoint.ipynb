```python
# 7자리 이상의 임의의 정수를 입력받은 후, 
# 1000단위 구별자를 추가하여 리턴하는 코드를 작성하시오...
# 123456789 => 123,456,789
```

```python
value = input("정수입력: ")
rest = len(value)%3
first = ""
print(rest)
if rest > 0:
    first = "value[:rest],"
num = 0
last = ""
for i in value[rest:]:
    num += 1
    if num >= 3:
        num = 0
        last += i+ ","
    else:
        last += i
print(first+last)
```

```python
num = input("임의의 7자리 이상인 정수를 입력하세요")

num = num[::-1]
ret = ''

for i, c in enumerate(num):
    
    i = i+1
    if(i%3 == 0) & (i != len(num)):
        ret += c + ','
    else:
        ret += c
        
ret = ret[::-1]

print(ret)
```

```python
# text file을 열어서 어떤 문자가 몇 번 등장헀는지를 출력
# 등장 빈도순으로 출력..
#'가' -> 256회
#'1' -> 230회
#'마' -> 150회
```

```python
with open("text.txt", 'r', encoding = "utf-8") as file:
    text = file.read()
    text_save = {}
    for i in text:
        if i in text_save.keys():
            text_save[i] += 1
        else:
            text_save[i] = 1
    sort_text = sorted(text_save.items(), key=lambda x: -x[1])
    for i in sort_text[1:]:
        print(f"{i[0]} => {i[1]}")
```
